```python
import requests
from bs4 import BeautifulSoup
import numpy as np

URL = "https://www.bayt.com/en/egypt/jobs/sales-jobs/"
page = requests.get(URL , headers={'User-Agent':'Mozilla/5.0'})
soup = BeautifulSoup(page.content, "html.parser")
```

```python
job_titles = []
job_titles = soup.find_all("h2",class_="jb-title m0 t-large")
```

```python
for titles in job_titles:
    print(titles.text)
```

```python
job_links=[]
for i in range(len(job_titles)):
    job_links.append('https://www.bayt.com'+job_titles[i].find('a').attrs['href'])
```

```python
for x in job_links :
    print(x)
```

```python
job_location = []
job_location = soup.find_all("span",class_="jb-loc t-mute t-nowrap")
```

```python
for location in job_location:
    print(location.text)
```

```python
company_name = []
company_name = soup.find_all("b", class_="jb-company")
```

```python
for name in company_name:
    print(name.text)
```

```python
job_description = []
job_description = soup.find_all("div",class_="jb-descr m10t t-small")
for desc in job_description:
    print(desc.text)
```

```python
date = []
date = soup.find_all("div", class_="jb-date col p0x t-xsmall t-mute")
for dt in date:
    print(dt.text)
```
