```python
import requests
from bs4 import BeautifulSoup
import numpy as np

URL = "https://www.bayt.com/en/egypt/jobs/sales-jobs/"
page = requests.get(URL , headers={'User-Agent':'Mozilla/5.0'})
soup = BeautifulSoup(page.content, "html.parser")
```

```python
job_titles = []
job_title = soup.find_all("h2",class_="jb-title m0 t-large")
```

```python
for titles in job_title:
    print(titles.text)
    job_titles.append(titles.text.split())
```

```python
job_links=[]
for i in range(len(job_title)):
    job_links.append('https://www.bayt.com'+job_title[i].find('a').attrs['href'])
```

```python
for x in job_links :
    print(x)
```

```python
job_locations = []
job_location = soup.find_all("span",class_="jb-loc t-mute t-nowrap")
```

```python
for location in job_location:
    print(location.text.split())
    job_locations.append(location.text.split())
```

```python
company_names = []
company_name = soup.find_all("b", class_="jb-company")
```

```python
for name in company_name:
    print(name.text)
    company_names.append(name.text.split())
```

```python
job_descriptions = []
job_description = soup.find_all("div",class_="jb-descr m10t t-small")
for desc in job_description:
    print(desc.text)
    job_descriptions.append(desc.text.split())
```

```python
dates = []
date = soup.find_all("div", class_="jb-date col p0x t-xsmall t-mute")
for dt in date:
    print(dt.text)
    dates.append(dt.text.split())
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(job_titles, job_links , job_locations , company_names , dates)),
               columns =['job title', 'job link', 'job location','company name' , 'date'])
```
